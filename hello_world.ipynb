# Hello World with py5

A simple introduction to py5 in Jupyter

````python
# This notebook uses the "py5" kernel (imported mode):
# py5 is pre-imported for you — no `import py5`, no `py5.` prefix,
# and Java 17 + the macOS event loop are already set up.
print("Hello, py5!")
````

````python
%%py5draw 400 300
# %%py5draw renders the sketch INLINE as an image (no window).
# This is the reliable way to show a sketch in Jupyter on macOS.
# The "%%py5draw W H" line MUST be the first line of the cell.
background(255)
fill(0, 100, 255)
rect(50, 50, 100, 100)
fill(255, 100, 0)
circle(300, 150, 50)
````

## Want a live window or an animation?

`run_sketch()` opens a separate native window. On **macOS this does not work from a Jupyter notebook** — the window can't be created from the kernel, so the cell just stalls.

For a windowed / animated sketch, save it as a `.py` file and run it from the Terminal:

```bash
JAVA_HOME=/usr/local/opt/openjdk@17 ~/miniconda3/bin/python my_sketch.py
```

Inside a notebook, stick with `%%py5draw` for static images (cell above). For an animation displayed inline, py5 also offers `py5_tools.sketch_portal()`.
